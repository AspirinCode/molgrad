```python
import os
os.chdir('../..')
import numpy as np
import pickle
from tqdm import tqdm
```

```python
from molexplain.vis import molecule_importance
from molexplain.utils import DATA_PATH, MODELS_PATH
from molexplain.train import DEVICE
from molexplain.prod import predict_mol
```

```python
import torch
model_pt = os.path.join(MODELS_PATH, 'cyp_noHs.pt')

from molexplain.net import MPNNPredictor

model = MPNNPredictor(node_in_feats=49,
                      edge_in_feats=10,
                      global_feats=4,
                      n_tasks=1,
                      output_f=torch.sigmoid).to(DEVICE) 
model.load_state_dict(torch.load(model_pt,
                                 map_location=DEVICE))
```

```python
from rdkit.Chem import MolFromSmiles, MolToInchi
import pandas as pd

sar1 = pd.read_csv(os.path.join(DATA_PATH, 'cyp', 'cyp_sar1.csv'), sep=';')  ## nothing interesting
sar1
```

```python
mol = MolFromSmiles(sar1['SMILES'][0])
print(predict_mol(mol, model))
_, img, _, _, global_importance = molecule_importance(mol,
                                                      model,
                                                      task=0,
                                                      version=2,
                                                      vis_factor=1,
                                                      eps=1e-4,
                                                      addHs=False)
img
```
