```python
from rdkit import Chem
from rdkit.Chem.Draw import IPythonConsole
```

```python
patt = Chem.MolFromSmarts('C-C-C-C-C(-[OH])=O')
patt
```

## Select carboxylic acids with aliphatic chain of 4 carbons or more

```python
smi = 'OC(=O)CCCCC1=CC=CC=C1'
mol = Chem.MolFromSmiles(smi)
print(mol.HasSubstructMatch(patt))
matches = mol.GetSubstructMatches(patt)
mol
```

```python
smi = 'CCC(C)C(C)CCC(O)=O'
mol = Chem.MolFromSmiles(smi)
print(mol.HasSubstructMatch(patt))
matches = mol.GetSubstructMatches(patt)
mol
```

## Do not select carboxylic acids with non-aliphatic (e.g. aromatic) carbons in their chain

```python
smi = 'CCC1=CC=CC(=C1)C(O)=O'
mol = Chem.MolFromSmiles(smi)
print(mol.HasSubstructMatch(patt))
matches = mol.GetSubstructMatches(patt)
mol
```

## Do not select esters

```python
smi = 'CCC(C)C(C)CCC(=O)OC'
mol = Chem.MolFromSmiles(smi)
print(mol.HasSubstructMatch(patt))
matches = mol.GetSubstructMatches(patt)
mol
```
