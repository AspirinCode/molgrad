```python
patterns = a
def is_fatty_acid(mol): 
    bools = [mol.HasSubstructMatch(patt) for patt in patterns]
    if True in bools: 
        if len([1 for atom in mol.GetAtoms() if atom.GetSymbol() != "C"]) == 2: 
            return True # fatty acid if we match the pattern and have only C and 2 oxygens
    return False
```

## Select carboxylic acids with aliphatic chain of 4 carbons or more

```python
smi = 'OC(=O)CCCCC1=CC=CC=C1'
mol = Chem.MolFromSmiles(smi)
bools = [mol.HasSubstructMatch(patt) for patt in patterns]
if is_fatty_acid(mol): 
    idx = len(patterns)-1-bools[::-1].index(True)
    matches = mol.GetSubstructMatch(patterns[idx])
print(is_fatty_acid(mol))
mol
```
