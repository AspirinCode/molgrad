## Scale and distribution of the endpoints

See the manuscript draft for details on where these were taken from.

```python
import os
os.chdir('../..')

import pickle
import numpy as np
import matplotlib.pyplot as plt
from molgrad.utils import DATA_PATH
```

```python
ENDPOINTS = ['ppb', 'herg', 'caco2']

f, axs = plt.subplots(1, 3, figsize=(12, 4))

for idx, end in enumerate(ENDPOINTS):
    with open(os.path.join(DATA_PATH, end, f'data_{end}.pt'), 'rb') as handle:
        _, values = pickle.load(handle)
        
        axs[idx].hist(values, bins=30)
        axs[idx].set_title(f'Endpoint {end}')
```
